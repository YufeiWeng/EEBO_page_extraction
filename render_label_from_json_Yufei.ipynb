```python
import json
from PIL import Image, ImageDraw

# Assuming `jsons_list` and `record` are defined somewhere in your code

data_path = "/trunk2/datasets/danlu/arabic_ms_data"
des = "/trunk2/yufei/datasets/labels/prediction"
os.makedirs(des, exist_ok=True)

for json_path in jsons_list:
    file_name = json_path.split("/")[-1].strip(".json")
    file_name = json_path.split("/")[-2] + "_" + file_name
    print("File name: ", file_name)

    img_path = record[json_path]
    size = Image.open(img_path).size
    new_canvas = Image.new("L", size)
    draw = ImageDraw.Draw(new_canvas)
    
    with open(json_path, 'r') as f:
        text_line_json = json.load(f)
    
    for item in text_line_json['text_line']:
        poly = item['polygon']
        flatten_line = []
        for point in poly:
            flatten_line.extend(point)  # Note the change here to accommodate the list of lists
        draw.polygon(flatten_line, fill=255)
    
    save_name_label = os.path.join(des, file_name + ".jpg")
    print(save_name_label)
    new_canvas.save(save_name_label)
```
